```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_temps = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
c_temps = [((f - 32) * 0.56) for f in f_temps]

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
plt.plot(months, f_temps)
plt.plot(months, c_temps)
plt.xlabel("Months")
plt.ylabel("Avg Temps")
plt.show()
```
