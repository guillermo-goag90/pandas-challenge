```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail_gun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
x_axis = np.arange(0, 130, 10)
x_axis
```

```python
plt.hlines(50, 0, 120, alpha=0.25)
plt.plot(x_axis, danger_drop, marker="o", color="blue")
plt.plot(x_axis, rail_gun, marker="^", color="red")
plt.title("Two Rollercoasters")
plt.xlabel("Time(s)")
plt.ylabel("Speeds")
plt.grid()
```
